```python
import pandas as pd

df = pd.read_csv('서울교통공사_역별 화재용 대피마스크 현황_20250310.csv', encoding='cp949')
df
```

```python
df['역명'] = df['역명'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
df['역명'] = df['역명'].str.replace(r'역$', '', regex=True).str.strip()

# 역명 기준으로 수량 합산
grouped = df.groupby('역명', as_index=False)['수량'].sum().sort_values('수량').rename(columns={'수량': '화재용 긴급대피마스크_수량'}).reset_index(drop=True)
grouped
```

```python
grouped.to_csv('마스크.csv', encoding='cp949', index=False)
```
