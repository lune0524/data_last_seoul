```python
df = pd.read_csv('서울교통공사_소화기설치현황_20250310.csv', encoding='cp949')

# 역명에서 괄호 제거하여 정제
df['역명'] = df['역명'].str.replace(r'\(.*\)', '', regex=True)

# 역명별로 보유대수 합계 구하기
df_sum = df.groupby('역명')['보유대수'].sum().reset_index().sort_values('보유대수')
df_sum.columns = ['역명', '소화기_수량']
df_sum
```

```python
station_df = pd.read_csv("서울교통공사_역사건축정보_20250310.csv", encoding="cp949")
station_area = station_df[['역명', '면적']]

# 3. 병합 (역명 기준)
df_merge = pd.merge(df_sum, station_area, on='역명', how='left')
df_merge['소화기_밀도(개/㎡)'] = df_merge['소화기_수량'] / df_merge['면적']
df_merge
```

```python
df_sum.to_csv('소화기.csv', encoding='cp949', index=False)
```
