1.데이터 불러오기

```python
import pandas as pd

# 파일 경로 목록
file_paths = [
    "서울교통공사_역별 시간대별 승하차인원(20.1~20.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(21.1~21.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(22.1~22.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(23.1~23.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(24.1~24.12).csv"
]

# 인코딩 문제 방지를 위해 cp949 사용
df_list = [pd.read_csv(fp, encoding='cp949') for fp in file_paths]

# 데이터프레임 이어붙이기
df = pd.concat(df_list, ignore_index=True)
df
```

```python
df.info()
```

```python
df = df.drop(['연번', '날짜', '역번호'], axis=1)
df
```

```python
df['호선'] = df['호선'].replace({1: '1호선', 2: '2호선', 3: '3호선', 4: '4호선', 5: '5호선', 6: '6호선', 7: '7호선', 8: '8호선'})
df.value_counts('호선')
```

```python
# 시간대(승-하차 인원) 컬럼만 추출
time_cols = df.columns.difference(
    ["호선", "역명", "구분"]
)

# 두 행씩 묶을 ‘pair’ 번호 부여
df["pair"] = df.index // 2          # 0,0,1,1,2,2 …

# 승차·하차 합계 계산
sum_df = df.groupby("pair")[time_cols].sum()

# 날짜·역명 등 메타데이터는 짝수행(승차)에서만 가져옴
meta = (
    df.loc[df.index % 2 == 0, ["호선", "역명", "구분"]]
      .set_index(df.loc[df.index % 2 == 0, "pair"])
)

# 합치고 정리
df = pd.concat([meta, sum_df], axis=1).reset_index(drop=True)
df = df.drop(columns=["구분"], axis=1)

df
```

```python
clos = df.columns[5:]
df['하루_유동인구'] = df[clos].sum(axis=1)
df
```

# 2. 역별 하루 평균 유동인구수(호선 구분 X)

```python
# 시간대 컬럼 추출
time_columns = df.columns[2:-1]

# 평균 계산
time_avg = df.groupby('역명')[time_columns].mean()
daily_avg = df.groupby('역명')['하루_유동인구'].mean().rename('하루_평균_유동인구')

# 호선 병합 (쉼표로)
line_merged = df.groupby('역명')['호선'].apply(lambda x: ','.join(sorted(x.astype(str).unique())))

# 통합
result_name = pd.concat([line_merged, time_avg, daily_avg], axis=1).reset_index()

# '호선' 컬럼을 따로 꺼내고 삭제 후 맨 앞에 삽입
line_col = result_name.pop('호선')      # 컬럼을 꺼내고 원본에서 제거
result_name.insert(0, '호선', line_col)  # 맨 앞(0번째)에 삽입

result_name
```

# 3. 호선별 하루평균 유동인구수(호선구분O)

```python
# 시간대 컬럼 추출 (예: '06시 이전' ~ '24시 이후')
time_columns = df.columns[2:-1]

# (역명, 호선) 기준으로 groupby
grouped = df.groupby('호선')

# 시간대별 평균 계산
time_avg = grouped[time_columns].mean()

# 하루 평균 유동인구 계산
daily_avg = grouped['하루_유동인구'].mean().rename('하루_평균_유동인구')

# 두 결과 통합
result_line = pd.concat([time_avg, daily_avg], axis=1).reset_index()
result_line
```

# 2-(1) 시각화

```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
plt.rcParams['axes.unicode_minus'] = False

# 그래프 그리기 - 하루 평균 유동인구
plt.figure(figsize=(8, 5))
plt.bar(result_line['호선'].astype(str), result_line['하루_평균_유동인구'])

plt.title('호선별 하루 평균 유동인구')
plt.xlabel('호선')
plt.ylabel('하루 평균 유동인구')
plt.show()
```

```python
# 시간대 컬럼 추출
time_columns = result_line.columns[1:-1]  # '06시 이전' ~ 마지막 시간대

# 그래프 그리기 - 시간대별 평균 유동인구
plt.figure(figsize=(10, 6))

for _, row in result_line.iterrows():
    plt.plot(time_columns, row[time_columns], label=row['호선']) 

plt.title('호선별 시간대별 평균 유동인구')
plt.xlabel('시간대')
plt.ylabel('평균 유동인구')
plt.legend(title='호선')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
