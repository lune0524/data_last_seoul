1.데이터 불러오기

```python
import pandas as pd

# 파일 경로 목록
file_paths = [
    "서울교통공사_역별 시간대별 승하차인원(20.1~20.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(21.1~21.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(22.1~22.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(23.1~23.12).csv",
    "서울교통공사_역별 시간대별 승하차인원(24.1~24.12).csv"
]

# 인코딩 문제 방지를 위해 cp949 사용
df_list = [pd.read_csv(fp, encoding='cp949') for fp in file_paths]

# 데이터프레임 이어붙이기
df = pd.concat(df_list, ignore_index=True)
df
```

```python
df.info()
```

```python
df = df.drop(['연번', '날짜', '역번호'], axis=1)
df
```

```python
df['호선'] = df['호선'].replace({1: '1호선', 2: '2호선', 3: '3호선', 4: '4호선', 5: '5호선', 6: '6호선', 7: '7호선', 8: '8호선'})
df.value_counts('호선')
```

```python
# 시간대(승-하차 인원) 컬럼만 추출
time_cols = df.columns.difference(
    ["호선", "역명", "구분"]
)

# 두 행씩 묶을 ‘pair’ 번호 부여
df["pair"] = df.index // 2          # 0,0,1,1,2,2 …

# 승차·하차 합계 계산
sum_df = df.groupby("pair")[time_cols].sum()

# 날짜·역명 등 메타데이터는 짝수행(승차)에서만 가져옴
meta = (
    df.loc[df.index % 2 == 0, ["호선", "역명", "구분"]]
      .set_index(df.loc[df.index % 2 == 0, "pair"])
)

# 합치고 정리
df = pd.concat([meta, sum_df], axis=1).reset_index(drop=True)
df = df.drop(columns=["구분"], axis=1)

df
```

```python
clos = df.columns[5:]
df['하루_유동인구'] = df[clos].sum(axis=1)
df
```

# 2. 역별 하루평균 유동인구수


```python
df_name_mean = df.groupby('역명')['하루_유동인구'].mean().reset_index()
df_name_mean = df_name_mean.rename(columns={'하루_유동인구': '역별_하루_평균유동인구'})

df_name_mean
```

```python
df_name_mean.to_csv('역별_하루_평균유동인구(2020~2024).csv', index=False, encoding='cp949')
```
